```python
import matplotlib.pyplot as plt
from datetime import datetime
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
days = 90
coin=[{'id':'hive','name':'HIVE','colour':'red'},
      {'id':'hive_dollar','name':'HBD','colour':'purple'},
      {'id':'splinterlands','name':'SPS','colour':'green'},
      {'id':'steem','name':'STEEM','colour':'blue'},
      {'id':'cub-finance','name':'Cub Finance','colour':'yellow'},
      {'id':'blurt','name':'Blurt','colour':'black'},
      {'id':'dark-energy-crystals','name':'DEC','colour':'magenta'}
     ]
r=[]
for c in coin:
    r.append(cg.get_coin_market_chart_by_id(id=c['id'], vs_currency='usd', days=days))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,10)
for i in range(len(r)):
    ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in r[i]['prices']], 
        [x[1] for x in  r[i]['prices']], label=coin[i]['name'], color=coin[i]['colour'])
ax.set_ylabel('Price (US$)')
ax.grid()
ax.legend();
```
