```python
import matplotlib.pyplot as plt
from datetime import datetime
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
days = 90
hivemc = cg.get_coin_market_chart_by_id(id='hive', vs_currency='usd', days=days)
hbd = cg.get_coin_market_chart_by_id(id='hive_dollar', vs_currency='usd', days=days)
steemmc = cg.get_coin_market_chart_by_id(id='steem', vs_currency='usd', days=days)
cub = cg.get_coin_market_chart_by_id(id='cub-finance', vs_currency='usd', days=days)
spsmc = cg.get_coin_market_chart_by_id(id='splinterlands', vs_currency='usd', days=days)
blurt = cg.get_coin_market_chart_by_id(id='blurt', vs_currency='usd', days=days)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,6)
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in hivemc['prices']], 
        [x[1] for x in  hivemc['prices']], label='HIVE', color='red')
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in hbd['prices']], 
        [x[1] for x in  hbd['prices']], label='HBD', color='purple')
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in spsmc['prices']], 
        [x[1] for x in  spsmc['prices']], label='SPS', color='green')
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in steemmc['prices']], 
        [x[1] for x in  steemmc['prices']], label='STEEM', color='blue')
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in cub['prices']], 
        [x[1] for x in  cub['prices']], label='Cub Finance', color='cyan')
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in blurt['prices']], 
        [x[1] for x in  blurt['prices']], label='Blurt', color='black')
ax.set_ylabel('Price (US$)')
ax.grid()
ax.legend();
```
