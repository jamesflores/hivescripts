```python
fromdate = (datetime.date.today()-datetime.timedelta(days=31)).strftime('%Y-%m-%d')
vtr, ptr, ttl = 0,0,0
fnd=[]
print('|Account|Location|Posting|Voting|')
print('|-|-|-|-|')
for s in searches:
    r = pd.read_sql(f"""select a.name, a.post_count, JSON_VALUE(posting_json_metadata,'$.profile.location') as loc,
                a.last_vote_time, a.last_post
                from Accounts a
                where a.posting_json_metadata like '%"location":%' and {s}
                    and (a.last_vote_time  >= '{fromdate}' or a.last_post  >= '{fromdate}')
                    order by a.name""", cnxn)
    for i, l in r.iterrows():
        if l['name'] not in fnd:
            flag = '*' if l['name'] in hw else ''
            ttl += 1
            poster = l['last_post'].strftime('%Y-%m-%d') >= fromdate
            if poster:
                ptr += 1
            voter = l['last_vote_time'].strftime('%Y-%m-%d') >= fromdate
            if voter:
                vtr += 1
            #print(f"@{l['name']} - {l['loc']} {flag} Posting: {poster} Voting: {voter}") #". Last post: {l['last_post']} Last vote: {l['last_vote_time']}")
            print(f"|@{l['name']} {flag}|{l['loc']}|{'*' if poster else ''}|{'*' if voter else ''}|")
            fnd.append(l['name'])
print(f'Total: {ttl} Posting: {ptr} Voting: {vtr}')
```
